```python
import numpy as np
import h5py
with h5py.File('/data/users/kyle/Mishas_Data/M83_h5/M83_abs.h5', 'r') as f:
    data_abs = f['rawdata']
    keys_abs = (list(data_abs.keys()))
    x_abs = np.empty([1100000,84])
    
    for i,n in enumerate(keys_abs):
        temp = data_abs.get(n)
        x_abs[:,i] = np.asarray(temp).flatten()
```

```python
freq_range = []    
for i in range(0,200):
    freq_range.append(1449.680/1e3-i*781.250/1e6)
```

```python
amp = np.transpose(x_abs[:,0].reshape(5500,200))
```

```python
import matplotlib.pyplot as plt
for xe, ye in zip(freq_range, amp):
    plt.scatter([xe] * len(ye), ye,c='black',s=0.2)
plt.show()
```
