```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
import h5py
with h5py.File('/data/users/kyle/Mishas_Data/M83_stokes_h5/4mag.h5', 'r') as f:
    x  = f['.']['features'].value
    
with h5py.File('/data/users/kyle/Mishas_Data/M83_stokes_h5/orig_flag.h5', 'r') as f:
    y  = f['.']['labels'].value
```

# Correlations between polarizations of a single baseline

```python
pols = pd.DataFrame({'XX':x[:,0,0], 'XY':x[:,1,0], 'YX':x[:,2,0], 'YY':x[:,3,0]})
corr = pols.corr()
corr.style.background_gradient(cmap='coolwarm')
```

# Correlations between polarizations without RFI of a single baseline

```python
pols_norfi = pd.DataFrame({'XX':x[y[:,0]==0,0,0], 'XY':x[y[:,0]==0,1,0], 'YX':x[y[:,0]==0,2,0], 'YY':x[y[:,0]==0,3,0]})
corr = pols_norfi.corr()
corr.style.background_gradient(cmap='coolwarm')
```

# Correlations between RFI in polarizations of a single baseline

```python
pols_rfi = pd.DataFrame({'XX':x[y[:,0]==1,0,0], 'XY':x[y[:,0]==1,1,0], 'YX':x[y[:,0]==1,2,0], 'YY':x[y[:,0]==1,3,0]})
corr = pols_rfi.corr()
corr.style.background_gradient(cmap='coolwarm')
```
